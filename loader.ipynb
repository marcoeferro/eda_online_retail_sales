```python
users = pd.read_csv("./data/cleaned/user_table.csv", encoding='latin1')
users.head(2)
```

```python
for index,row in users.iterrows():
    id1 = row["user_id"]
    name1 = row["user_name"]
    insert_user(id1,name1)
```

```python
products = pd.read_csv("./data/cleaned/product_table.csv", encoding='latin1')
products.head(2)
```

```python
for index, row in products.iterrows():
    product_id = row["product_id"]
    product_name = row["product_name"]
    category_1 = row["category_1"]
    category_2 = row["category_2"]
    category_3 = row["category_3"]
    category_4 = row["category_4"]
    category_5 = row["category_5"]
    discounted_price = row["discounted_price"]
    actual_price = row["actual_price"]
    discount_percentage = row["discount_percentage"]
    rating = row["rating"]
    rating_count = row["rating_count"]
    about_product = row["about_product"]
    img_link = row["img_link"]
    product_link = row["product_link"]
    insert_product(product_id,product_name,category_1,category_2,category_3,category_4,category_5,discounted_price,actual_price,discount_percentage,rating,rating_count,about_product,img_link,product_link)
```

```python
review = pd.read_csv("./data/cleaned/review_table.csv", encoding='latin1')
review.head(2)
```

```python
for index, row in review.iterrows():
    review_id = row["review_id"]
    review_title = row["review_title"]
    review_content = row["review_content"]
    review_user_id = row["user_id"]
    review_product_id = row["product_id"]
    insert_review(id=review_id,id_user=review_user_id,id_product=review_product_id,title=review_title,content=review_content)
```
