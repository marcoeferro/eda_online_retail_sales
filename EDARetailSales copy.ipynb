# Productos mas Vendidos

```python
grouped_df = df.groupby(["id"]).size().reset_index(name='count')

# Sort by count in descending order
grouped_df.sort_values(by='count', ascending=False, inplace=True)

# Get the most popular product's information
fst_prod = grouped_df.iloc[0]
scd_prod = grouped_df.iloc[1]
thr_prod = grouped_df.iloc[2]
frd_prod = grouped_df.iloc[3]
fth_prod = grouped_df.iloc[4]
```

```python
product_ids = [fst_prod.id, scd_prod.id, thr_prod.id,frd_prod.id,fth_prod.id]
top5_df = df.loc[df["id"].isin(product_ids),["id", "product_name", "category_1", "category_2", "category_3", "category_4", "category_5", "discounted_price", "actual_price", "discount_percentage", "rating", "rating_count", "about_product", "img_link", "product_link"]].drop_duplicates(subset=["id"])
```
